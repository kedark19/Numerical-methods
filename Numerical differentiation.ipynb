```python
#!usr/bin/env python3

# euler'method :differntial equation

import math
import numpy as np

def f(y,t):
   return t*y+ pow(t,3)

t=np.linspace(0,1,1000)
s=0

h=t[1]-t[0]
m=len(t)
y=np.zeros(m)
y[0]=1

for i in range(m-1):
  y[i+1]=y[i]+ h*f(y[i],t[i])
  


import matplotlib.pyplot as plt
plt.plot(t,y,'bs')        # plot
plt.show()
```
